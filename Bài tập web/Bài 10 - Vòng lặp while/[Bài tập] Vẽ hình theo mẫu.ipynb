```python
import turtle
import random
t = turtle.Turtle()
turtle.bgcolor("black")
t.pensize(2)
i = 0
CL = ['pink', 'blue', 'green', 'yellow', 'orange', 'cyan', 'purple', 'red']

while i < 36:
    t.color(random.choice(CL))
    t.circle(50,90)
    t.circle(100,90)
    t.circle(50,90)
    t.circle(100,90)
    t.rt(10)
    i += 1
```
